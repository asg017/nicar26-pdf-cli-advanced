```python
import pdfplumber
```

Pdf Plumber relies on text and line information embeddded in the PDF to extract content. Since we OCRed our PDF, PDF plumber does not know where to look. We have to add some lines to guide the extraction.

```python
pdf = pdfplumber.open("lasd-restraining-orders_ocr_basic.pdf")
# print(pdf.pages[0].extract_text())
```

```python
p0 = pdf.pages[0]
```

```python
# p0.to_image().draw_rects(p0.extract_words())
```

```python
table_data = pdf.pages[0].extract_table({
    "vertical_strategy": "explicit",
    "horizontal_strategy": "text",
    "explicit_vertical_lines": [50,110,350,436,538,584,826,866,905,944]
})
```

```python
for row in table_data:
    print('\t'.join(row))
```
